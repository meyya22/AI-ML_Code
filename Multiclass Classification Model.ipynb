In the given training dataset, there are 5 classes attributed by various factors. The objective of this excercise is to evaluate the F1 score by using the train data (700 rows) and test data (300 rows), also export the result in CSV format

```python
import pandas as pd
df = pd.read_csv("C:/Users/Dell/Desktop/train.csv")
df.head(5)
```

Identfy the significant variables/columns and check the mean different. Ignore the variable for this excercise if the mean difference is not significant for the given 5 class (0,1,2,3,4)

```python
df1 = df[['category_class','sold_price','price','size']]
df1.groupby('category_class').describe()
```

Feature Extraction and Data mapping for sting values (area_name and condition)

```python
df_dummies_c = pd.get_dummies(df.condition, prefix="condition")
df_with_dummies_c = pd.concat([df,df_dummies_c],axis='columns')
df_with_dummies_c.head()
```

```python
df_dummies = pd.get_dummies(df.area_name, prefix="area_name")
df_with_dummies = pd.concat([df_with_dummies_c,df_dummies],axis='columns')
df_with_dummies.head()
```

```python
df_with_dummies.drop('item_id',axis='columns',inplace=True)
df_with_dummies.drop('condition',axis='columns',inplace=True)
df_with_dummies.drop('listing_at',axis='columns',inplace=True)
df_with_dummies.drop('item_tag_hash',axis='columns',inplace=True)  
df_with_dummies.drop('area_name',axis='columns',inplace=True) 
df_with_dummies.drop('category_class',axis='columns',inplace=True) 
df_with_dummies.head()
```

```python
X = df_with_dummies
```

```python
y = df.category_class
```

Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.429)
```

Choose a model which gives better F1 score - LogisticRegression / SVM / DT

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
df_test = pd.read_csv("C:/Users/Dell/Desktop/test.csv")
df_test.head()
```

Preparing Test Data
